```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
%matplotlib inline
```

```python
df = pd.read_csv("/content/Sample - Superstore.csv", encoding="latin1")

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
customer_sales = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False)

customer_sales.head(10)
```

```python
plt.figure(figsize=(12,6))

customer_sales.head(10).plot(kind="bar", color="orange")

plt.title("Top 10 Customers by Sales")
plt.ylabel("Sales")
plt.xlabel("Customer Name")

plt.show()
```

```python
customer_profit = df.groupby("Customer Name")["Profit"].sum().sort_values(ascending=False)

customer_profit.head(10)
```

```python
plt.figure(figsize=(12,6))

customer_profit.head(10).plot(kind="bar", color="green")

plt.title("Top 10 Customers by Profit")
plt.ylabel("Profit")
plt.xlabel("Customer Name")

plt.show()
```

```python
customer_orders = df.groupby("Customer Name")["Order ID"].count().sort_values(ascending=False)

customer_orders.head(10)
```

```python
plt.figure(figsize=(12,6))

customer_orders.head(10).plot(kind="bar", color="purple")

plt.title("Top Customers by Number of Orders")
plt.ylabel("Orders")

plt.show()
```

# Business Insights

- A small number of customers generate a significant portion of sales.
- Some customers contribute high profits, making them valuable for retention.
- Frequent customers tend to place multiple orders throughout the year.
- Identifying top customers helps businesses design loyalty programs.
- Customer segmentation enables targeted marketing and personalized promotions.
